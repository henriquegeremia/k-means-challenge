```python
# done
def df_to_points(df):
    '''
    for each row of the dataframe is created a vector to represent a point
    '''
    n_of_registers, n_of_features = df.shape
    
    points = []
    for i in range(0,n_of_registers-1):
        serie = df.iloc[i]
        vector = pd.Series.tolist(serie)
        points.append(vector)
    return points
```
